```python
import numpy as np
```

## Примеры создания массивов

```python
lst = [i for i in range(20)] # Создание списка из чисел от 0 до 19 при помощи списочного выражения
arr = np.array(lst) # Список становится одномерным массивом numpy
```

```python
print(arr)
```

```python
zero_arr = np.zeros((5, 5)) # Создание квадратной нулевой матрицы размером 5
```

```python
zero_arr
```

```python
ones_arr = np.ones((2, 4)) # Создание матрицы, заполненной единицами, размером 2 на 4
```

```python
ones_arr
```

```python
arange_arr = np.arange(10).reshape(5, 2) # Создание одномерной матрицы с элементами от 0 до 9, затем ее превращение в матрицу 5 на 2
```

```python
arange_arr
```

```python
diag = np.eye(3) # Создание диагональной матрицы размером 3
```

```python
diag
```

## Примеры индексирования и срезов

```python
arange_arr
```

```python
arange_arr[1, 0] # Элемент массива, находящийся в первой строке, нулевом столбце
```

```python
arange_arr[2:, 1:] # Подмассив (строки, начиная со второй, столбцы, начиная с первого)
```

## Примеры векторных операций и базовых статистик

```python
arange_arr * arange_arr # Поэлементное умножение массивов
```

```python
arr
```

```python
arr_mean = np.mean(arr) # Среднее массива

arr_min = np.min(arr) # Мин. элемент массива
arr_max = np.max(arr) # Макс. элемент массива
norm_arr = (arr - arr_min) / (arr_max - arr_min) # Нормализация данных

print(f'Среднее массива: {arr_mean}\nМакс. элемент массива: {arr_max}\nМин. элемент массива: {arr_min}\nНормализованный массив: {norm_arr}')
```
